```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the 'scrape' column to string type
df['scrape'] = df['scrape'].astype(str)

# Set the colors for the bars
colors = {'0': 'salmon', '1': 'turquoise'}

# Create the count plot
sns.countplot(data=df, x='scrape', palette=colors)

# Set plot labels and title
plt.xlabel('Scrape Value')
plt.ylabel('Count')
plt.title('Histogram of Scrape Column')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the 'scrape' column to string type
df['scrape'] = df['scrape'].astype(str)

# Set the colors for the bars
colors = {'0': 'salmon', '1': 'turquoise'}

# Create the count plot
ax = sns.countplot(data=df, x='scrape', palette=colors)

# Annotate the bars with the count
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12, color='black')

# Set plot labels and title
plt.xlabel('Scrape Value')
plt.ylabel('Count')
plt.title('Histogram of Scrape Column')

# Show the plot
plt.show()
```

```python
# Filter the DataFrame to include only rows where 'scrape' is 1
final_scrape = df[df['scrape'] == '1']  # Assuming 'scrape' is a string in this example

# Write the filtered DataFrame to a CSV file
final_scrape.to_csv('sites_to_scrappe.csv', index=False)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(final_scrape):
    # Drop columns: 'comments', 'Website Link'
    final_scrape = final_scrape.drop(columns=['comments', 'Website Link'])
    # Rename column 'update_link' to 'link'
    final_scrape = final_scrape.rename(columns={'update_link': 'link'})
    return final_scrape

final_scrape_clean = clean_data(final_scrape.copy())
final_scrape_clean.head()
```
