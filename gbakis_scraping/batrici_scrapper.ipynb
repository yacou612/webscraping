```python
##scrapper 1: page vente

import requests
from bs4 import BeautifulSoup
import csv

# URL of the webpage to scrape
url = 'https://batirici-immobilier.com/vente/'

# Send a GET request to the webpage
response = requests.get(url)
# Parse the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the property listings
property_listings = soup.find_all('div', class_='property-unit-information-wrapper')

# List to store the scraped data
properties = []

# Extract the required information from each listing
for listing in property_listings:
    price = listing.find('div', class_='listing_unit_price_wrapper').get_text(strip=True).split()[0]
    agent = listing.find('div', class_='property_location').get_text(strip=True)
    
    beds = listing.find('span', class_='inforoom')
    beds = beds.get_text(strip=True) if beds else 'N/A'
    
    baths = listing.find('span', class_='infobath')
    baths = baths.get_text(strip=True) if baths else 'N/A'
    
    size = listing.find('span', class_='infosize')
    size = size.get_text(strip=True).split()[0] if size else 'N/A'
    
    details = listing.find('div', class_='listing_details the_list_view')
    details = details.get_text(strip=True) if details else 'N/A'
    
    properties.append([price, agent, beds, baths, size, details])

# Define the CSV file header
header = ['Price', 'Agent', 'Beds', 'Baths', 'Size_m2', 'Details']

# Write the data to a CSV file
with open('batrici_vente.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(properties)

print("Data has been written to batrici_vente.csv")
```
